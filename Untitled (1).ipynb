```python
title=[]
ratings=[]
price=[]
```

```python
def extract_record(item):
    """Extract and return the Data from a Single record"""
    
    #Description
    atag = item.h2.a
    description = atag.text.strip()
    
    try:
        #Price
        price_parent = item.find('span', 'a-price')
        price = price_parent.find('span', 'a-offscreen').text
    except AttributeError: #Attribute error handeling
        price = ''
    
    try:
        #Rating
        rating = item.i.text
    except AttributeError: #Attribute error handeling
        rating = ''
        
        
    result = (description, price, rating)
    
    return result
```

```python
records = []
soup = BeautifulSoup(driver.page_source, 'html.parser')
results = soup.find_all('div',{'data-component-type' : 's-search-result'})

for item in results:
    record = extract_record(item)
    if record:
        records.append(record)
        
records = records[:10]

driver.close()

title =[]
Ratings =[]
Price =[]

for i in records:
    title.append(i[0])
    Ratings.append(i[2])
    Price.append(i[1])
    
intel_core_i7_amazon = pd.DataFrame({"Title":title,"Ratings":Ratings,"Price":Price})
intel_core_i7_amazon
```

```python
# Send an HTTP request to the web page
url = "https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table element
table = soup.find("table")

# Extract data from the table
data = []
for row in table.find_all("tr"):
    cells = row.find_all("td")
    if len(cells) > 0:
        data.append([cell.text.strip() for cell in cells])

df = pd.DataFrame(data, columns=['No','PM Name', 'Born-Dead', 'Term of office', 'Remark'])

# Print the dataframe
print(df)
```

```python
# Ques 8
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
chn_tab=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
chn_tab.click()
```

```python
quote = []
author = []
type_of_quote = []
```

```python
j = 0
while j != 10:
    quo=driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quo:
        q=i.text
        quote.append(q)

    auth=driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in auth:
        a=i.text
        author.append(a)
    
    types=driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in types:
        t=i.text
        type_of_quote.append(t)
        
    next_tab=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a")
    next_tab.click()
    
    j += 1

        
        
    
```

```python
print(len(quote),len(author),len(type_of_quote))
```

```python
import pandas as pd
df=pd.DataFrame({'Author':author,'Quotes':quote,'Types_of_Quote':type_of_quote})
df
```
