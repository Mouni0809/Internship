```python
#Ques 1

url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python
#Scraping Rank
rank_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
try:
    for i in rank_tag[0:30]:
        rank.append(i.text)
except NoSuchElementException:
    rank.append("-")
except ElementNotVisibleException:
    rank.append("--")
    
#Scraping Name
name_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
try:
    for i in name_tag[0:30]:
        v_name=i.text.split('[')
        name.append(v_name[0])
except NoSuchElementException:
    name.append("-")
except ElementNotVisibleException:
    name.append("--")
    
#Scraping Artist
artist_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
try:
    for i in artist_tag[0:30]:
        artist.append(i.text)
except NoSuchElementException:
    artist.append("-")
    
#Scraping Upload Date
date_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
try:
    for i in date_tag[0:30]:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append("-")
    
#Scraping Views
view_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
try:
    for i in view_tag[0:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append("-")
```

```python
mv=pd.DataFrame({'Rank':rank,'Video Name':name,'Artist':artist,'Upload Date':upload_date,'Views':views})
mv
```

```python
#Ques 4

url = 'https://github.com/'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
```

```python
open_source = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
try:
    open_source.click()
    time.sleep(5)
except ElementNotInteractableException:
    driver.get(open_source.get_attribute('href'))
```

```python
trend = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
try:
    driver.get(trend.get_attribute('href'))
except ElementNotInteractableException:
    driver.get(trend.get_attribute('href'))
```

```python
title= []
description= []
count= []
language= []
urls=[]
```

```python
#Scraping title
try:
    rt=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
    for i in rt:
        title.append(i.text)
except:
    title.append('-')
    
#Scraping description 
try:
    desc=driver.find_elements(By.XPATH,'//article[@class="Box-row"]/p')
    for i in desc:
        description.append(i.text)
except:
    description.append('-')

for i in [rt,desc]:
    for j in i:
        if i ==rt:
            title.append(j.text)
            urls.append(j.get_attribute('href'))
        if i==desc:
            description.append(j.text)
    
for url in urls:
    driver.get(url)
    driver.implicitly_wait(3)
    dlist=driver.find_elements(By.XPATH,'//div[@class="BorderGrid BorderGrid--spacious"]/div')
    #Scraping contributors   
    try:
        count.append(((dlist[-2].text).split())[1])
    except:
        count.append('-')
    #Scraping language
    try:
        language.append(((dlist[-1].text).split())[1::2])
    except:
        language.append('-')
```

```python
trending=pd.DataFrame({"Repository_title":title[:25],"Repository_description":description[:25],"Language_used":language[:25],"Contributors_count":count[:25]})
trending
```

```python
#Scraping name
Name_tag=driver.find_elements(By.XPATH,"//h3[@class='c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet']")
for i in Name_tag:
    Name.append(i.text)
    
nameTag=driver.find_elements(By.XPATH,"//h3[@class='c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only']")
for i in nameTag:
    Name.append(i.text )

#Scraping artist
Artist_tag=driver.find_elements(By.XPATH,"//span[@class='c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only u-font-size-20@tablet']")
for i in Artist_tag:
    Artist.append(i.text)
    
artistTag=driver.find_elements(By.XPATH,"//span[@class='c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only']")
for i in artistTag:
    Artist.append(i.text )


#Scraping rank
rankTag=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-bold-l a-font-primary-m@mobile-max u-font-weight-normal@mobile-max lrv-u-padding-tb-050@mobile-max u-font-size-32@tablet']")
for i in rankTag[:3]:
    rank.append(i.text)
    
RankTag=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in RankTag:
    Rank.append(i.text)
        
# slicing Data
for i in Rank:
    if i=='':
        Rank.remove(i)
        
last_week_rank=Rank[0::3]
last_week_rank.insert(0,rank[0])
peak_rank=Rank[1::3]
peak_rank.insert(0,rank[1])
weeks_on_board=Rank[2::3]
weeks_on_board.insert(0,rank[2])
```

```python
songs_top=pd.DataFrame({"Song Name":Name[:100],"Artist Name":Artist[:100],"Last Week Rank":last_week_rank[0:100],"Peak Rank":peak_rank[0:100],"Weeks On Board":weeks_on_board[0:100]})
songs_top
```

```python
#Ques 8

url = 'https://archive.ics.uci.edu/'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
```

```python
dataset=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
dataset.click()
time.sleep(3)
```

```python
name = []
types = []
task = []
attribute_type = []
attribute_count = []
instance = []
year = []
```

```python
j = 0

while j <= 64:   
    time.sleep(2)
    
    exp=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]')
    exp.click()
    time.sleep(3) 
        
    #Scraping Name
    dname=driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a')
    try:
        for i in dname:
            name.append(i.text)
    except:
        name.append("--") 
        
    #Scraping Types    
    dtype=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
    try:
        for i in dtype:
            types.append(i.text)
    except:
        types.append("--") 
        
    #Scraping Task    
    dtask=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
    try:
        for i in dtask:
            task.append(i.text)
    except:
        task.append("--") 
        
    #Scraping Attribute    
    a_type=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
    try:
        for i in a_type:
            if i == " ":
                pass
            else:
                attribute_type.append(i.text)
    except:
        attribute_type.append("--") 
        
    #Scraping Instance    
    ins=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
    try:
        for i in ins:
            instance.append(i.text)
    except:
        instance.append("--") 
        
    #Scraping Attribute count    
    a_count=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
    try:
        for i in a_count:
            attribute_count.append(i.text)
    except:
        attribute_count.append("--") 
        
    #Scraping Year    
    dat=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
    try:
        for i in dat:
            if i == " ":
                pass
            else:
                year.append(i.text)
    except:
        year.append("--")   
        
        
    exp2=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]')
    exp2.click()
    time.sleep(2)
    
    driver.execute_script("window.scrollBy(0,10000)")
       
    nexttab=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    nexttab.click()   
    j += 1
```

```python
Datasets= pd.DataFrame({"Dataset name":name,"Data type":types,"Task":task,"Attribute type":attribute_type,"Instances":instance,"No of Attribute":attribute_count,"Year":year})
Datasets
```
